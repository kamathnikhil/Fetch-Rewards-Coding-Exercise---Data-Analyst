In this python file, I will convert the JSON file to CSV format. Before that I will clean and format the data so that I can use it for data processing and modeling. 

```python
#I am using basic pandas and datetime libraries to get started. I will add more libraries if required
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
#info() will give me details of the data like, datatype, count and number of columns
#head will show me top 10 rows of the df
receipts = pd.read_json('receipts.json.gz',lines=True)
receipts.info()
receipts.head(10)
```

```python
#This gives you a clear on percentage of values missing in a specific attribute or column
percent_missing = receipts.isnull().sum() * 100 / len(receipts)
missing_value_receipts = pd.DataFrame({'column_name': receipts.columns,
                                 'percent_missing': percent_missing})
```

```python
missing_value_receipts
```

```python
''' I need to convert the date format for more than 1 field therfore creating a function 
 where X is string in UTC and Y is the datetime format '''
def date_converter(x):
    try:
        return(datetime.utcfromtimestamp(int(x['$date'])/1000).strftime('%Y-%m-%d %H:%M:%S'))
    except TypeError:
        return(None)
```

```python
#using lambda to separate oid and value key. I just need value key i.e. 5ff1e1d20a7214ada1000561
receipts['_id'] = receipts['_id'].apply(lambda x: x['$oid'])
receipts = receipts.rename(columns={'_id': 'id'})
receipts['createDate'] = receipts['createDate'].apply(lambda x: date_converter(x))
receipts['dateScanned'] = receipts['dateScanned'].apply(lambda x: date_converter(x))
receipts['finishedDate'] = receipts['finishedDate'].apply(lambda x: date_converter(x))
receipts['modifyDate'] = receipts['modifyDate'].apply(lambda x: date_converter(x))
receipts['pointsAwardedDate'] = receipts['pointsAwardedDate'].apply(lambda x: date_converter(x))
receipts['purchaseDate'] = receipts['purchaseDate'].apply(lambda x: date_converter(x))
```

```python
#changing the datatype for the date field
receipts['purchaseDate'] = pd.to_datetime(receipts['purchaseDate'])
receipts['pointsAwardedDate'] = pd.to_datetime(receipts['pointsAwardedDate'])
receipts['modifyDate'] = pd.to_datetime(receipts['modifyDate'])
receipts['finishedDate'] = pd.to_datetime(receipts['finishedDate'])
receipts['dateScanned'] = pd.to_datetime(receipts['dateScanned'])
receipts['createDate'] = pd.to_datetime(receipts['createDate'])
```

#### rewardsReceiptItemList has 18 columns. Therefore, I have extracted the column and made a separate data. I can use this data to join receipt data so that I can use the attributes in rewardsReceiptItemList data

```python
#for loop for extracting rewardsReceiptItemList column out of receipts table
l = []
for i in range(len(receipts)):
    try:
        l.append(receipts['rewardsReceiptItemList'].iloc[i][0])
    except:
        #print(receipts['rewardsReceiptItemList'].iloc[i])
        pass
```

```python
#separating rewardreceiptitemlist from the table, so we will need a primary key and its attribute
rewards_receipts_itemlist = pd.DataFrame(l)
rewards_receipts_itemlist["rewardsReceiptStatus"] = receipts["rewardsReceiptStatus"]
rewards_receipts_itemlist['receipts_id']= receipts["id"] # copied receipts ID to the new table for joining and as there is 1 to many relationship between the table
rewards_receipts_itemlist.sample()
```

```python
# Generating unique identifier for 'rewards_receipts_itemlist' table
rewards_receipts_itemlist['receipts_item_id'] = np.zeros((len(rewards_receipts_itemlist),1))
for i in range(len(rewards_receipts_itemlist)):
    rewards_receipts_itemlist['receipts_item_id'].iloc[i]= i+1
```

```python
#we dont need rewardsReceiptItemList in receipts data
receipts=receipts.drop(['rewardsReceiptItemList'], axis=1)
```
